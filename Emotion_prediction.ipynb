```python
# Load the saved pipeline and model
import joblib
import pandas as pd
from sklearn import preprocessing
from sklearn.metrics import precision_score, recall_score, f1_score
saved_file = open("/content/drive/-------/weight/", "rb")
trained_model = joblib.load(saved_file)
```

```python
def predict_on_narnia(df, trained_model):
    df = df.rename(columns={"sent":"Text"})
    df = df.rename(columns={"annotate1":"Emotion"})
    # making suprise negative and suprise positive as a single suprise emotion
    df.loc[:, 'Emotion'] = df.loc[:, 'Emotion'].replace('Su-', 'Su')
    df.loc[:, 'Emotion'] = df.loc[:, 'Emotion'].replace('Su+', 'Su')
    X = df['Text']
    # X = df['Text'].apply(preprocessing)
    y_true = df['Emotion']

    # Predict emotions
    y_pred = trained_model.predict(X)

    # print the results
    # Loop through the dataframe and print actual emotion, predicted emotion, and sentence
    for idx, (actual, predicted, sentence) in enumerate(zip(y_true, y_pred, df['Text'])):
        print(f"Example {idx+1}:")
        print(f"Actual Emotion: {actual}")
        print(f"Predicted Emotion: {predicted}")
        print(f"Sentence: {sentence}\n")

    # Calculate precision, recall, and F1-score
    precision = precision_score(y_true, y_pred, average='weighted')
    recall = recall_score(y_true, y_pred, average='weighted')
    f1 = f1_score(y_true, y_pred, average='weighted')

    # Print the results
    print(f'Precision: {precision:.4f}')
    print(f'Recall: {recall:.4f}')
    print(f'F1-Score: {f1:.4f}')

# Assuming df is your DataFrame and pipe_lr is your pipeline
# Call the function like this:
```

```python
df = pd.read_csv('/content/drive/MyDrive/Text emotion detection/Dataset/test/NarniyaThe LastBattle01102023.csv')
```

```python
predict_on_narnia(df, trained_model)
```
